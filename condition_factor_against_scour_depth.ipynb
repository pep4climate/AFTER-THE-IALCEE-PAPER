Linear elastic segmental masonry arch with fixed ends subjected to vertical loads and displacement

```python
results_dict = {"delta_v": [0, 0.10, 0.50],
                "reaction_x_sx": [143.30, 143.30, 143.30],
                "reaction_z_sx": [306.70, 269.50, 120.80],
                "moment_y_sx": [19.30, 200.30, 924.10],
                "reaction_x_dx": [-143.30, -143.30, -143.30],
                "reaction_z_dx": [306.70, 343.90, 492.70],
                "moment_y_dx": [-19.3, 161.70, 885.50]}
```

```python
results_df = pd.DataFrame.from_dict(results_dict)
```

```python
results_df
```

```python
results_df_thrust = results_df
```

```python
# https://sparkbyexamples.com/pandas/iterate-over-rows-in-pandas-dataframe/
# https://pythonspeed.com/articles/pandas-vectorization/
results_df_thrust["reaction_thrust"] = results_df["reaction_x_sx"] * math.sin(alfa_rad) + results_df["reaction_z_sx"] * math.cos(alfa_rad)
```

```python
results_df_thrust
```

```python
results_df_thrust["scour_depth_dimensionless"] = results_df_thrust["delta_v"]/foundation_depth
results_df_thrust["reaction_thrust_dimensionless"] = results_df_thrust["reaction_thrust"]/max(results_df_thrust["reaction_thrust"])
```

```python
plt.scatter(results_df_thrust["scour_depth_dimensionless"], results_df_thrust["reaction_thrust_dimensionless"], color='k')
plt.xlabel("pier vertical displacement/foundation depth (-)", fontSize=14)
plt.ylabel("dimensionless thrust reaction (-)", fontSize=14)
plt.xticks(np.arange(0, 1+0.1, 0.1), fontSize=14)
plt.yticks(np.arange(0, 1+0.1, 0.1), fontSize=14)
```

```python
# https://www.statology.org/matplotlib-trendline/
x = results_df_thrust["scour_depth_dimensionless"]
y = results_df_thrust["reaction_thrust_dimensionless"]

#create scatterplot
plt.scatter(x, y, color='k')

#calculate equation for trendline
z = np.polyfit(x, y, 1)
p = np.poly1d(z)

#add trendline to plot
plt.plot(np.arange(0, 1+0.1, 0.1), p(np.arange(0, 1+0.1, 0.1)), color='k')

plt.xlabel("pier vertical displacement/foundation depth (-)", fontSize=14)
plt.ylabel("dimensionless thrust reaction (-)", fontSize=14)
plt.xticks(np.arange(0, 1+0.1, 0.1), fontSize=14)
plt.yticks(np.arange(0, 1+0.1, 0.1), fontSize=14)
```

```python
# https://stackoverflow.com/questions/48126428/how-to-make-the-inverse-of-a-linear-functionhttps://stackoverflow.com/questions/48126428/how-to-make-the-inverse-of-a-linear-function
relative_scour_depth_failure = (0-z[1]/z[0])
relative_scour_depth_failure
```

```python
#counter check
p(relative_scour_depth_failure)
```

```python
# https://www.statology.org/matplotlib-trendline/
x = results_df_thrust["scour_depth_dimensionless"]
y = results_df_thrust["reaction_thrust_dimensionless"]

#create scatterplot
plt.scatter(x, y, color='k')

#calculate equation for trendline
z = np.polyfit(x, y, 1)
p = np.poly1d(z)

#add trendline to plot
plt.plot(np.arange(0, 1+0.1, 0.1), p(np.arange(0, 1+0.1, 0.1)), color='k')

plt.xlabel("pier vertical displacement/foundation depth (-)", fontSize=14)
plt.ylabel("dimensionless thrust reaction (-)", fontSize=14)
plt.xticks(np.arange(0, 1+0.1, 0.1), fontSize=14)
plt.yticks(np.arange(0, 1+0.1, 0.1), fontSize=14)

plt.scatter(relative_scour_depth_failure, p(relative_scour_depth_failure), color='orange')
x=round(relative_scour_depth_failure, 2)
y=round(p(relative_scour_depth_failure), 2)
zip([x], [y]) #https://realpython.com/python-zip-function/#using-zip-in-python
label = f"({x},{y})"
plt.annotate(label,
                (x,y),
                textcoords="offset points",
                xytext=(-40, -10),
                ha='center',
                rotation=0,
                fontSize=14)

plt.xticks(np.arange(0, 1+0.1, 0.1), fontSize=14)
plt.yticks(np.arange(0, 1+0.1, 0.1), fontSize=14)

legend_labels = ["linear elastic double fixed", "obtained function", "relative scour depth failure probability"]
plt.legend(legend_labels, loc='best')
```
