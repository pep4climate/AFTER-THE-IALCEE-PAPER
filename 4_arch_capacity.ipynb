```python
import matplotlib.pyplot as plt
import numpy as np
import scipy
```

```python
%store -r arch_capacity_variables
```

```python
print(arch_capacity_variables)
```

```python
b=arch_capacity_variables["b"] #https://www.w3schools.com/python/python_dictionaries_access.asp
t=arch_capacity_variables["t"]
e=arch_capacity_variables["e"]
```

```python
%store -r r
```

```python
Rc=r
```

```python
condition_factor = 0.9
```

```python
deterministic_coefficient = 0.6
```

```python
arch_capacity = deterministic_coefficient * condition_factor * (Rc*1000*b*(t-2*e)) #https://en.wikipedia.org/wiki/Normal_distribution#Operations_on_a_single_normal_variable
arch_capacity[0:5]
```

```python
arch_capacity_mean = np.mean(arch_capacity)
print(arch_capacity_mean)
```

```python
arch_capacity_std = np.std(arch_capacity)
print(arch_capacity_std)
```

```python
arch_capacity_realizations = np.linspace(arch_capacity_mean-3*arch_capacity_std, arch_capacity_mean+3*arch_capacity_std,100)
```

```python
arch_capacity_distribution=scipy.stats.norm.pdf(arch_capacity_realizations, arch_capacity_mean, arch_capacity_std)
```

```python
plt.plot(arch_capacity_realizations, arch_capacity_distribution, color='black')
plt.xlabel("Compressive resistance force")
plt.ylabel("Probability density")
```
